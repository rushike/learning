## Spectral lines csv to elements.spectral_lines.parquet table


```python
import pandas

df = pandas.read_csv("../data/spectral-lines.csv")

sprectral_lines = df.groupby(["element"]).agg({"nm" : list})
sprectral_lines.index = sprectral_lines.index.rename("symbol")
sprectral_lines = sprectral_lines.rename(columns = {"nm" : "spectral_lines"})



sprectral_lines.to_parquet("../tables/elements.spectral_lines.parquet")
sprectral_lines
```

## Elements Json to all elements data table

```python
import pandas 

df = pandas.read_json("../data/elements.json")
df = df.set_index("symbol")
df = df.join(sprectral_lines)
# df = df.rename(columns= {"wavelength": "spectral_lines"})
df.to_parquet("../tables/elements.all.parquet")
df
```
